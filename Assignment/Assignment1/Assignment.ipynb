Assignment 1: Temperature Check 

```python
temperature=float(input("Enter temperature in Celsius: "))
if temperature<10:
    print("It's too Cold")
elif temperature>=10 and temperature<=25:
    print("It's cool outside")
else:
    print("It's Hot outside")
```

Assignment 2: ATM Withdrawal System 

```python
print("Enter your account balance:")
balance=float(input())
print("Enter the withdrawal amount:")
withdrawal=float(input())
if withdrawal > balance:
    print("Insufficient Balance")
else:
    print("Transaction Successful")
```

Assignment 3: Student Grade Calculator 

```python
print("Enter Your marks")
marks=float(input())

if marks>=90 and marks<=100:
    print("Your Grade is A+")
elif marks>=80 and marks<90:
    print("Your Grade is A")
elif marks>=70 and marks<80:
    print("Your Grade is B")
elif marks>=60 and marks<70:
    print("Your Grade is C")
else: 
    print("Your Grade is F")
```

Assignment 4: Bus Ticket Discount

```python
print("Enter your age:")
age=int(input())
if age<5:
    print("Ticket is Free")
elif age>=5 and age<=18:
    print("You have to pay 50% of the ticket price")
elif age>60:
    print("You have to pay 70% of the ticket price")
else:
    print("You have to pay full ticket price")
```

Assignment 5: Restaurant Bill with Discount

```python
amount=int(input("Enter the amount: "))
if amount>=1000:
    discount=amount*0.1
    final_amount=amount-discount
    print("Final bill amount.:",final_amount)
else:
    print("No discount")
```

Assignment 7: Traffic Light System

```python
color=input("Enter Traffic LIght color: (Red/Yellow/Green) ")
if color.lower()=='red':
    print("Stop")
elif color.lower()=='yellow':
    print("Ready to go")
elif color.lower()=='green':
    print("Go")
else:
    print("Invalid Color")
```

Assignment 8: Leap Year Checker 

```python
year=int(input("Enter the year: "))
if(year%400==0):
    print("leap year")
elif(year%4==0 and year%100!=0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

Assignment 10: Electricity Bill Calculator

```python
unit=float(input("Enter the unit consumed: "))
if unit<=100:
    bill=unit*5
elif unit>100 and unit<=200:
    bill=100*5+(unit-100)*7
else:
    bill=100*5+100*7+(unit-200)*10
    
print("total payable amount.:",bill)
```
